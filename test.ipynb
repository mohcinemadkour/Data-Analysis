```python
def plot_roc(list_df,colors,list_labels,list_AUC,name,cutoffs,x=10, y=5.7):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    for df,c,l,auc,cutoff in zip(list_df,colors,list_labels,list_AUC,cutoffs):
        ax.plot(1-df['specificity'],df['sensitivity'],color=c,label=l+'\n AUC '+auc,linewidth=2)
        cut=df[df['thres']==cutoff]
        ax.plot(1-cut['specificity'],cut['sensitivity'],c+'D',label='',markersize=8)
    ax.plot(1-df['specificity'],1-df['specificity'],color='k',label='',linewidth=3)
    extra = Rectangle((0, 0), 1, 1, fc="w", fill=False, edgecolor='none', linewidth=0)
    handles, labels = ax.get_legend_handles_labels()
    ax.grid()
    ax.legend([extra]+handles, [u"\u2666" + ' Cutoff']+labels,loc=4)
    
    plt.xlabel('1-Specificity')
    plt.ylabel('Sensitivity')
    plt.title(' ROC curve for '+ name)
    ax2=plt.gcf().set_size_inches(x, y)
    return plt
```

```python
def plot_performance_measure(list_df,list_labels,cutoffs,x=10, y=5.7):
    n=len(list_labels)
    fig,axes = plt.subplots(nrows=n/2, ncols=2, figsize=(x*(2.1), y*(n/2+.2)))

    for df,l,c,ax in zip(list_df,list_labels,cutoffs,axes.flat):
        ax.plot(df['thres'],df['acc'],'b',label='Accuracy')
        ax.plot(df['thres'],df['ppv'],'r',label='Positive Predictive Value (PPV)')
        ax.plot(df['thres'],df['npv'],'g',label='Negative Predictive Value (NPV)')
        ax.plot(df['thres'],df['yod_index'],'c',label='Youden Index')
        ax.plot([c,c],[0,1],'k:',label='')
        ax.grid()
        ax.legend(loc=5)
        ax.set_xlabel('Performance')
        ax.set_ylabel('Threshold')
        ax.set_title(l)
        
    fig.tight_layout()    

    return fig
```

```python
def plot_cutoffs_roc(list_df,colors,list_labels,cutoffs,x=10, y=5.7):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    for df,c,l,cutoff in zip(list_df,colors,list_labels,cutoffs):
        ax.plot(df['thres'],df['yod_index'],color=c,label=l+'\n Cutoff '+str(cutoff),linewidth=2)
        cut=df[df['thres']==cutoff]
        ax.plot(cut['thres'],cut['yod_index'],c+'D',label='',markersize=8)
    #ax.plot(1-df['specificity'],1-df['specificity'],color='k',label='',linewidth=3)
    extra = Rectangle((1, 1), 0, 0, fc="w", fill=False, edgecolor='none', linewidth=0)
    handles, labels = ax.get_legend_handles_labels()
    ax.grid()
    ax.legend([extra]+handles, [u"\u2666" + ' Cutoff']+labels,loc=4,bbox_to_anchor=(1.0,0.5))
    
    
    plt.xlabel('threshold')
    plt.ylabel('yoden index')
    plt.title(' Decision Criteria (cuttofs) using Training cohort ')
    ax2=plt.gcf().set_size_inches(x, y)
    return plt
```
